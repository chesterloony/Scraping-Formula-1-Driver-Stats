Importing dataset from motosportstats.com.

```python
import requests
import json
import pandas as pd

url = "https://www.motorsportstats.com/api/results-summary?seasonSlug=formula-one_2022&seriesSlug=formula-one&stats=championshipWin,championshipRank,starts,wins,podiums,poles,fastestLaps,bestFinishPosition,bestGridPosition,retirements,avgFinishPosition,avgGridPosition,racesLed,pointsFinishes,points&size=25"

payload={}
headers = {}

r = requests.get(url, headers=headers)

driverdata = r.json()
```

```python
driverdata.keys()
```

```python
driverdata['content'][0]
```

```python
df = pd.json_normalize(driverdata['content'])
```

```python
print(df.head())
```

Cleaning the dataset.

```python
df = df[["driver.name", "team.code", "stats.poles", "stats.fastestLaps", "stats.bestFinishPosition", "stats.bestGridPosition", "stats.retirements", "stats.avgFinishPosition", "stats.avgGridPosition", "stats.racesLed", "stats.pointsFinishes", "stats.points"]]
```

```python
df
```

```python
df = df.rename(columns={"driver.name": "Name", "team.code": "Team", "stats.poles": "Pole Positions", "stats.fastestLaps": "Fastest Laps", "stats.bestFinishPosition": "Best Finish Position", "stats.bestGridPosition": "Best Grid Position", "stats.retirements": "Retirements", "stats.avgFinishPosition": "Avg Finish Position", "stats.avgGridPosition": "Avg Grid Position", "stats.racesLed": "Races Led", "stats.pointsFinishes": "Points Finishes", "stats.points": "Points"})
```

```python
df.to_csv('driverdata2022.csv', index=False)
```
